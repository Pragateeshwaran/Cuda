## ‚öôÔ∏è GPU Execution Model

### üöÄ One Kernel Launch = One Grid  
Each time a **kernel** is launched, it creates and executes **one grid** of threads on the GPU.

If you launch **multiple kernels**, each kernel gets its **own grid**.

So you can think of grids as per-kernel, not per-GPU.


### Example
```cpp
kernelA<<<dim3(4,4), dim3(16,16)>>>();
// This launches ONE grid with 4x4 = 16 blocks, each with 16x16 = 256 threads

kernelB<<<dim3(2), dim3(128)>>>();
// Another kernel launch, another SINGLE grid, 2 blocks of 128 threads each
```

---

### üß† What is a Kernel?  
A **kernel** is a small, GPU-executable function written to run **in parallel**.  
It is executed **concurrently** by **thousands of lightweight threads**.
